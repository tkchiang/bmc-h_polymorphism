```python
import os
import numpy as np
from numpy.fft import fft2, fftshift
from tifffile import imread

from scipy.ndimage import gaussian_filter
from scipy.signal import find_peaks
from scipy.ndimage import rotate
from scipy.optimize import curve_fit

from tqdm import tqdm

import matplotlib.pyplot as plt
import pandas as pd
```

### Image directory and file names

```python
image_dir = '../../Image data/TEM/Nanotube TEM images/'
image_files = sorted([f for f in os.listdir(image_dir) if f.lower().endswith('.tif')])
```

### Load/process images and compute Fourier transforms

```python
# Function to get the center crop of an image
def get_center_crop(image, crop_size):
    h, w = image.shape
    crop_h = h // crop_size
    crop_w = w // crop_size
    start_row = int(round((h - crop_h) / 2))
    end_row   = int(round((h + crop_h) / 2))
    start_col = int(round((w - crop_w) / 2))
    end_col   = int(round((w + crop_w) / 2))
    image_cropped = image[start_row:end_row, start_col:end_col]
    return image_cropped

sigma = 30 # Low-pass filter cutoff in pixels

results = []
for fname in tqdm(image_files, desc="Processing images and computing FFTs"):

    path = os.path.join(image_dir, fname)
    im_raw = imread(path).astype(np.float64)

    # Subtract Gaussian-smoothed image
    im_smooth = im_raw - gaussian_filter(im_raw, sigma=sigma)

    # Compute 2D Fourier Transform
    ft2d = fftshift(fft2(im_smooth))

    # Store results
    results.append({
        'filename': fname,
        'im_raw': im_raw,
        'im_smooth': im_smooth,
        'ft2d': ft2d
    })
```

### Function to estimate tube principal axis angle

```python
def estimate_tube_orientation_pca(im, percentile_low=10, percentile_high=20):
    
    if not np.issubdtype(im.dtype, np.floating):
        im = im.astype(np.float64)

    # Flatten and find intensity threshold for the darkest pixels
    threshold_low = np.percentile(im, percentile_low)
    threshold_high = np.percentile(im, percentile_high)

    # Get (y, x) coordinates of pixels below the threshold
    y, x = np.where((im >= threshold_low) & (im <= threshold_high))
    coords = np.column_stack((x, y))

    # Center the coordinates
    coords_centered = coords - np.mean(coords, axis=0)

    # PCA via SVD
    _, _, vh = np.linalg.svd(coords_centered, full_matrices=False)
    direction = vh[0]

    # Compute angle (in degrees from horizontal axis)
    angle_rad = np.arctan2(direction[1], direction[0])
    angle_deg = np.rad2deg(angle_rad) % 180

    return angle_deg
```

### Function to get azimuthal Fourier profile and peaks

```python
def get_profiles(fourier_image, expected_feature_size=6.8/2, plot=False):
    
    pixelSize = 200 / 562  # nm per pixel
    Npixels = fourier_image.shape[0]
    image_center = Npixels / 2

    # Compute the range of q_values to profile based on expected feature size
    size_margin = 0.05 * expected_feature_size
    size_range = expected_feature_size + size_margin * np.array([-1, 1])
    expected_q = (crop_size * (Npixels * pixelSize / size_range)).astype(int)
    q_max, q_min = expected_q

    # Azimuthal angles to sample
    angles = np.linspace(0, 2*np.pi, 1000)

    # Get profile rows between q_min and q_max
    profile_rows = []
    for q in range(q_min, q_max + 1):
        qxy = np.round(q * np.stack([np.cos(angles), np.sin(angles)], axis=1) + image_center).astype(int)
        intensity_values = []
        for qx, qy in qxy:
            if 0 <= qy < Npixels and 0 <= qx < Npixels:
                intensity_values.append(fourier_image[qy, qx])
            else:
                intensity_values.append(0)
        intensity_values = np.array(intensity_values)
        mad = np.median(np.abs(intensity_values - np.median(intensity_values)))
        intensity_values = (intensity_values - np.median(intensity_values)) / mad
        profile_rows.append(intensity_values)

    profile_rows = np.array(profile_rows)
    
    # Get the maximum values from all profile rows
    profile = np.max(profile_rows, axis=0)

    # Find the 12 peaks in the azimuthal profile
    angle_spacing = np.mean(np.diff(angles))
    min_separation_rad = np.pi / 12
    min_separation_pts = int(np.round(min_separation_rad / angle_spacing))
    min_prom = 3*np.std(profile)

    # Extend profile by wrapping around. Deals with peaks that are too close to boundaries
    profile_extended = np.concatenate([profile[-min_separation_pts:], profile, profile[:min_separation_pts]])

    n_peaks = 0
    while n_peaks !=12:
        # Find peaks in the extended profile
        peak_inds_ext, _ = find_peaks(profile_extended, prominence=min_prom, distance=min_separation_pts)

        # Map back to original indices
        peak_inds = (peak_inds_ext - min_separation_pts) % len(profile)
        
        # Remove duplicates due to wrapping
        peak_inds = np.unique(peak_inds)
        n_peaks = len(peak_inds)
        min_prom = 0.99*min_prom

    # Measure the real-space feature size
    real_sizes = []
    for i in range(len(peak_inds)):
        radial_slice = profile_rows[:,peak_inds[i]]
        q_peak = np.argmax(radial_slice) + q_min - 1
        real_size = (crop_size * (Npixels * pixelSize / q_peak))
        real_sizes.append(real_size)

    real_sizes = np.array(real_sizes)

    # Optional plotting to check
    if plot:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(fourier_image, cmap='gray', origin='lower')

        plt.subplot(1, 2, 2)
        plt.plot(np.rad2deg(angles), profile)
        plt.plot(np.rad2deg(angles[peak_inds]), profile[peak_inds], 'rx')
        plt.show()

    return profile, profile_rows, angles, peak_inds, real_sizes, expected_q
```

### Get refined tube axis angle measurement and the chiral angle.

```python
def get_tube_axis_angle_and_chirality(angles, tube_angle_estimate_deg):

    raw_angles = np.mod(angles, 2 * np.pi)
    tube_angle_estimate = np.mod(np.deg2rad(tube_angle_estimate_deg), 2 * np.pi)

    # Angular distance accounting for periodicity
    angular_diffs = np.abs(np.angle(np.exp(1j * (raw_angles - tube_angle_estimate))))
    closest_two_inds = np.sort(np.argsort(angular_diffs)[:2])  # ensure increasing order
    i1, i2 = closest_two_inds

    closest_two_angles = raw_angles[closest_two_inds]
    new_tube_angle_estimate = np.mean(closest_two_angles)

    # Create angle sets using modulo for wraparound
    angle_set_1 = angles[np.mod(np.arange(i1, i1 + 12, 2), 12)]
    angle_set_2 = angles[np.mod(np.arange(i2, i2 + 12, 2), 12)]

    delta_degrees = np.abs(angle_diff_deg(angle_set_2, angle_set_1) / 2)
    alpha_avg = np.mean(delta_degrees)
    alpha_std = np.std(delta_degrees)

    return new_tube_angle_estimate, alpha_avg, alpha_std

# Compute difference between angles a2 and a1 in degrees, accounting for periodicity.
def angle_diff_deg(a2, a1):
    
    diff = np.rad2deg(a2 - a1)
    diff_wrapped = (diff + 180) % 360 - 180  # Wrap to [-180, 180)
    
    return diff_wrapped
```

### Function to get tube diameter and uncertainty

```python
def get_tube_width_and_uncertainty(im, tube_angle):
    pixelSize = 200 / 562  # nm per pixel

    angle_iter = 0
    while angle_iter < 12:
        test_angle = tube_angle + angle_iter * np.pi/6

        # Rotate raw image by a test angle
        im_rot = rotate(im, angle=np.rad2deg(test_angle), reshape=False, order=1, mode='constant', cval=0.0)

        # Get smoothed intensity profile and take the derivative to find tube edges
        intensity_profile = np.nanmean(np.where(im_rot == 0, np.nan, im_rot), axis=1)
        intensity_profile_smooth = gaussian_filter(intensity_profile, sigma=10)
        diff_profile = np.abs(np.diff(intensity_profile_smooth))
        std_diff = np.std(diff_profile)

        # Find peaks in difference profile corresponding to tube edges
        min_prom = 10 * std_diff
        numEdges = 0
        prom_iter = 0
        while numEdges != 2 and prom_iter < 200 and min_prom > 3*std_diff:
            tubeEdges, _ = find_peaks(diff_profile, prominence=min_prom, distance=100)
            numEdges = len(tubeEdges)
            min_prom *= 0.99
            prom_iter += 1

        if numEdges == 2:
            corrected_tube_angle = test_angle
            break
        else:
            angle_iter += 1

    # Fit diff_profile to 2-Gaussian function
    def fitDiffProfile(x, A1, A2, mu1, mu2, sigma1, sigma2):
        peak1 = A1 * np.exp(-0.5 * ((x - mu1) / sigma1) ** 2)
        peak2 = A2 * np.exp(-0.5 * ((x - mu2) / sigma2) ** 2)
        return peak1 + peak2

    mu0 = sorted(tubeEdges[:2])
    A0 = [diff_profile[mu] for mu in mu0]
    sigma0 = [1 for mu in mu0]
    offset0 = np.min(diff_profile)
    x_vals = np.arange(len(diff_profile))

    popt, _ = curve_fit(fitDiffProfile, x_vals, diff_profile, p0=[A0[0], A0[1], mu0[0], mu0[1], sigma0[0], sigma0[1]])

    width_pixels = np.abs(popt[2] - popt[3])
    width_pixels_err = np.sqrt(np.sum(popt[4:6]**2))
    width_nm = width_pixels*pixelSize
    width_nm_err = width_pixels_err*pixelSize
    tubeEdges = [popt[2], popt[3]]

    return width_nm, width_nm_err, width_pixels, width_pixels_err, tube_angle, tubeEdges
```

### Perform the batch analysis on all the nanotubes

```python
tube_angles = []
tube_widths = []
tube_widths_err = []
alphas = []
alphas_err = []

pixelSize = 200 / 562  # nm per pixel

doPlot = True

def imnorm(im):
    return (im - np.nanmean(im)) / np.nanstd(im)

for index in range(len(results)):
    filename = results[index]['filename']
    crop_size = 3.5
    im = results[index]['im_raw']
    ft2d = results[index]['ft2d']
    ft_crop = get_center_crop(np.abs(ft2d), crop_size)

    # Estimate tube axis angle
    axis_angle_estimate = estimate_tube_orientation_pca(im)

    # Get Fourier profile, peak angles, real sizes
    profile, profile_rows, angles, peak_inds, real_sizes, expected_q = get_profiles(ft_crop, expected_feature_size=6.8 / 2)

    # Refine axis angle measurement and get chiral angle
    tube_angle, alpha_avg, alpha_std = get_tube_axis_angle_and_chirality(angles[peak_inds], axis_angle_estimate)

    alphas.append(alpha_avg)
    alphas_err.append(alpha_std)

    # Measure tube width
    width_nm, width_nm_err, width_pixels, width_pixels_err, corrected_tube_angle, tubeEdges = get_tube_width_and_uncertainty(im, tube_angle)

    tube_widths.append(width_nm)
    tube_widths_err.append(width_nm_err)
    tube_angles.append(corrected_tube_angle)

    if doPlot and index % 30 == 0: # Show a few example results
        # Plotting for inspection
        plt.figure(figsize=(16, 6))
        
        # Raw image
        plt.subplot(1, 4, 1)
        plt.imshow(imnorm(im), cmap='gray', origin='lower')
        plt.title(filename)
    
        # Cropped FFT
        plt.subplot(1, 4, 2)
        plt.imshow(imnorm(ft_crop), cmap='gray', origin='lower')
        plt.title(f"FFT of {filename}")
    
        # Rotated image
        plt.subplot(1, 4, 3)
        im_rot = rotate(im, angle=np.rad2deg(corrected_tube_angle), reshape=False, order=1, mode='constant', cval=0.0)
        plt.imshow(imnorm(im_rot), cmap='gray', origin='lower')
        Npixels = im_rot.shape[0]
        x = [Npixels*0.25, Npixels*0.75]
        plt.plot(x, [np.mean(tubeEdges), np.mean(tubeEdges)], 'r-')
        for edge in tubeEdges:
            plt.plot(x, [edge, edge], 'b:', linewidth=1)
            plt.plot(x, [edge, edge] + width_pixels_err, 'r:', linewidth=1)
            plt.plot(x, [edge, edge] - width_pixels_err, 'r:', linewidth=1)
        plt.title(f"Width (nm): {np.round(width_nm, 2)} ± {np.round(width_nm_err, 2)}")
    
        # Rotated FFT
        plt.subplot(1, 4, 4)
        ft_rot = rotate(np.abs(ft2d), angle=tube_angle*180/np.pi, reshape=False, order=1, mode='constant', cval=0.0)
        ft_rot_crop = get_center_crop(ft_rot, crop_size)
        plt.imshow(imnorm(ft_rot_crop), cmap='gray', origin='lower')
        image_center = ft_rot_crop.shape[0] // 2
        plt.plot(image_center + np.array([-1, 1])*100, image_center + np.array([-1, 1]), 'r')
    
        grain_angle_1 = 90 - alpha_avg
        grain_angle_2 = 90 + alpha_avg
        for color, grain_angle in zip(['c', 'm'], [grain_angle_1, grain_angle_2]):
            grain_axis = np.array([np.cos(np.deg2rad(grain_angle)), np.sin(np.deg2rad(grain_angle))])
            plt.plot([image_center, image_center + grain_axis[0]*expected_q[0]/3], [image_center, image_center + grain_axis[1]*expected_q[0]/3], color)
        plt.title(f"α = {np.round(alpha_avg, 2)}° ± {np.round(alpha_std, 2)}°")
    
        plt.tight_layout()
        plt.show()
```

### Store results in `tube_results.csv`

```python
# Assuming the widths measured from TEM images are the width of a flattened nanotube, i.e. width = circumference / 2.
# Convert the width values to diameter00
tube_diameters = 2 * np.array(tube_widths) / np.pi
tube_diameters_err = 2 * np.array(tube_widths_err) / np.pi

df = pd.DataFrame({
    "Diameter": tube_diameters,
    "Alpha": alphas,
    "Diameter_err": tube_diameters_err,
    "Alpha_err": alphas_err
})

df.to_csv("tube_results.csv", index=False)
```
