```python
import os
import numpy as np

from tqdm import tqdm

import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
import pandas as pd
```

```python
# Import results from `nanotube_TEM_analysis.ipynb`
df = pd.read_csv("tube_results.csv")
```

### Display diameter/chiral angle scatterplot and histograms

```python
g = sns.JointGrid(data=df, x="Diameter", y="Alpha", height=6, space=0.1)
g.ax_joint.errorbar(
    df["Diameter"], df["Alpha"],
    xerr=df["Diameter_err"],
    yerr=df["Alpha_err"],
    fmt='o', markersize=2, elinewidth=0.5, capsize=0, alpha=0.5
)

# Diameter histogram
sns.histplot(data=df, x="Diameter", bins=30, ax=g.ax_marg_x, edgecolor=None)
g.ax_marg_x.set_ylabel("Counts")
g.ax_marg_x.tick_params(left=True, labelleft=True)

# Chiral angle histogram
sns.histplot(data=df, y="Alpha", bins=30, ax=g.ax_marg_y, edgecolor=None)
g.ax_marg_y.set_xlabel("Counts")
g.ax_marg_y.tick_params(bottom=True, labelbottom=True)
for ax in [g.ax_joint, g.ax_marg_x, g.ax_marg_y]:
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

g.set_axis_labels("Nanotube diameter (nm)", "Chiral angle (degrees)")
g.ax_joint.set_title(f'N = {len(df["Alpha"])}')

# Save figure output
#plt.savefig("Scatterplot_diameters_alphas.pdf")

plt.show()
```

### Get the chiral indices from nanotube diameters and chiral angles

```python
def get_chiral_indices(diameter, chiral_angle_deg):
    
    # Convert chiral angle from degrees to radians
    chiral_angle = np.deg2rad(chiral_angle_deg)

    # Initial guess for m and n
    min_diff = float('inf')
    m, n_final = None, None

    # Lattice constant
    a = 6.81

    # Iterate over possible values of m and n
    for m in range(1, 100):  # Iterate over possible m values
        for n in range(0, m):  # Constrain n to values less than m
            
            # Calculate the expected diameter for this (m, n)
            expected_diameter = (a / np.pi) * np.sqrt(m**2 + m*n + n**2)

            # Calculate the chiral angle for this (m, n)
            expected_angle = np.arctan(n / (m + n))
            expected_angle_deg = np.rad2deg(expected_angle)

            # Check if the diameter and chiral angle are close
            diameter_diff = abs(expected_diameter - diameter)
            angle_diff = abs(expected_angle_deg - chiral_angle_deg)

            # Find the closest (m, n) pair
            if diameter_diff < min_diff and angle_diff < min_diff:
                min_diff = diameter_diff
                m_final, n_final = m, n

    return m_final, n_final

# Get chiral indices (m, n)
M = np.zeros(len(df["Alpha"]))
N = np.zeros(len(df["Alpha"]))
for i in range(len(df["Alpha"])):
    m, n = get_chiral_indices(df["Diameter"][i], df["Alpha"][i])
    M[i] = m
    N[i] = n

df['M'] = M
df['N'] = N

maxMN = np.max((np.max(M).astype(int), np.max(N).astype(int)))
MN_heatmap = np.zeros((maxMN+2, maxMN+2))

for i in range(len(df["Diameter"])):
    MN_heatmap[(M[i]).astype(int), (N[i]).astype(int) ] += 1
```

### Display chiral index heatmap

```python
fig, ax = plt.subplots(figsize=(12, 6))
im = ax.imshow(MN_heatmap, cmap='BuPu', origin='lower')

ax.yaxis.set_inverted = 'False'
ax.set_xlabel("Chiral Index $n$")
ax.set_ylabel("Chiral Index $m$")
ax.set_xlim(-0.5, 32.5)
ax.set_ylim(19.5, 51.5)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="3%", pad=0.15)

cbar = plt.colorbar(im, cax=cax)
cbar.set_label('Counts')

# Save figure output
#plt.savefig("Heatmap_mn.pdf")

plt.show()
```
